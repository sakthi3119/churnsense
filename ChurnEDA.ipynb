# DATA CLEANING

```python
copy = base.copy()
```

```python
copy.TotalCharges = pd.to_numeric(copy.TotalCharges,errors='coerce')
copy.isnull().sum()
```

```python
copy.loc[copy['TotalCharges'].isnull()==True]
```

```python
copy.dropna(how='any',inplace=True)
```

```python
print(copy['tenure'].max())
```

## GROUPING THE TENURE

```python
labels = ["{0}-{1}".format(i,i+11) for i in range(1,72,12)]
copy['tenure_group'] = pd.cut(copy.tenure,range(1,80,12),right=False,labels=labels)
```

```python
copy['tenure_group'].value_counts()
```

```python
copy.drop(columns=['customerID','tenure'],axis=1,inplace=True)
copy.head()
```

# ACTUAL EDA BEGINS

### UNIVARIATE ANALYSIS

```python
for i,predictor in enumerate(copy.drop(columns=['Churn','TotalCharges','MonthlyCharges'])):
    plt.figure(i)
    sns.countplot(data=copy,x=predictor,hue='Churn')
    plt.show()
```

```python
copy['Churn']=np.where(copy.Churn=='Yes',1,0)
```

```python
copy.head()
```

```python
copy_dummies=pd.get_dummies(copy)
copy_dummies.head()
```

```python
sns.lmplot(data=copy_dummies,x='MonthlyCharges',y='TotalCharges',fit_reg=False)
plt.show()
```

### INSIGHT 3

--Total charges increase as Monthly Charges increases

```python
month = sns.kdeplot(copy_dummies.MonthlyCharges[(copy_dummies["Churn"]==0)],color="Red",fill=True)
month = sns.kdeplot(copy_dummies.MonthlyCharges[(copy_dummies["Churn"]==1)],ax=month,color="Blue",fill=True)

month.legend(["No Churn","Churn"],loc='upper right')
month.set_ylabel('Density')
month.set_xlabel('Monthly Charges')
month.set_title('Monthly charges by Churn')
plt.show()
```

### INSIGHT 4
-- Churn is high when Monthly Charges are high

```python
total = sns.kdeplot(copy_dummies.TotalCharges[(copy_dummies["Churn"]==0)],color="Red",fill=True)
total = sns.kdeplot(copy_dummies.TotalCharges[(copy_dummies["Churn"]==1)],ax=total,color="Blue",fill=True)

total.legend(["No Churn","Churn"],loc='upper right')
total.set_ylabel('Density')
total.set_xlabel('Total Charges')
total.set_title('Total charges by Churn')
plt.show()
```

### INSIGHT 5
-- Higher churn at lower Total Charges

```python
plt.figure(figsize=(20,8))
copy_dummies.corr()['Churn'].sort_values(ascending=False).plot(kind='bar')
plt.show()
```

### INSIGHT 6

-- High churn is observed in **Month to month contracts**, **No online security**, **No Tech support**, **First year of subscription** and **Fibre Optics Internet**.

-- Low churn is observed in **Long term contracts**, **Subscriptions without internet service** and **The customers engaged for 5+ years**

-- Factors like **Gender**, **Availability of PhoneService** and **no of multiple lines** have almost no impact on Churn

```python
plt.figure(figsize=(12,12))
sns.heatmap(copy_dummies.corr(),cmap="Paired")
plt.show()
```

## BIVARIATE ANALYSIS

```python
df_target0=copy.loc[copy["Churn"]==0]
df_target1=copy.loc[copy["Churn"]==1]
```

```python
def uniplot(df,col,title,hue=None):
    sns.set_style('whitegrid')
    sns.set_context('talk')
    plt.rcParams['axes.labelsize']=20
    plt.rcParams['axes.titlesize']=22
    plt.rcParams['axes.titlepad']=30

    temp = pd.Series(data=hue)
    fig,ax=plt.subplots()
    width = len(df[col].unique()) + 7 + 4 * len(temp.unique())
    fig.set_size_inches(width,8)
    plt.xticks(rotation=45)
    #plt.yscale('log')
    plt.title(title)
    ax=sns.countplot(data=df,x=col,order=df[col].value_counts().index,hue=hue,palette='bright')

    plt.show()
```

```python
uniplot(df_target1,col='Partner',title='Distribution of Gender for Churned Customers',hue='gender')
```

```python
uniplot(df_target0,col='Partner',title='Distribution of Gender for Non Churned Customers',hue='gender')
```

```python
uniplot(df_target1,col='PaymentMethod',title='Distribution of PaymentMethod for Churned Customers',hue='gender')
```

```python
uniplot(df_target1,col='Contract',title='Distribution of Contract for Churned Customers',hue='gender')
```

```python
uniplot(df_target1,col='TechSupport',title='Distribution of TechSupport for Churned Customers',hue='gender')
```

```python
uniplot(df_target1,col='SeniorCitizen',title='Distribution of SeniorCitizen for Churned Customers',hue='gender')
```
